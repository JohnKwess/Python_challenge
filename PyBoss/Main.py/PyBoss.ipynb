```python

digit = Data["SSN"].str.split("-")
digitlist = []
for row in digit :
    row[0] = "***-"
    row[1] ="**-"
    x = pd.Series([row[0], row[1], row[2]]).str.cat()
    digitlist.append(x)
Data["SSN"] = digitlist 
Data["DOB"] = pd.to_datetime(Data["DOB"]).dt.strftime('%m/%d/%Y')     


Data.head()
cols = Data.columns.tolist()
cols = cols[-2:] + cols[:-2]
Updatecsv = Data[cols].drop(columns = "Name")
```
